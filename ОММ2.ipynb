```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import math

pi = 3.1415926535
e = 2.7182818284
N = 90
M = 100
L = 101
hx = pi/(2*(N-1))
tau = 1/(10*(M-1))
hy = pi/(L-1)

y = np.zeros(L)
x = np.zeros(N)
t = np.zeros(M)
z = np.zeros((N,L,M))
w = np.zeros((N,L))
alpha_x = np.zeros(N)
bet_x = np.zeros(N)
alpha_y = np.zeros(L)
bet_y = np.zeros(L)

for i in range(N):
    x[i] = hx * i
    
for i in range(M):
    t[i] = tau * i
    
for i in range(L):
    y[i] = hy * i
    
for i in range(N):
    for k in range(L):
        z[i,k,0] = np.sin(2*x[i])*np.sin(3*y[k])
        
Ax = Bx = 9*tau/(2*hx**2)
Cx = 9*tau/hx**2 + 1

Ay = By = 9*tau/(2*hy**2)  
Cy = 9*tau/hy**2 + 1

for j in range(M-1):
    #вспомогательный слой
    for k in range(1,L-1):
        alpha_x[0] = 0
        bet_x[0] = 0
        for i in range(1,N-1):
            F = z[i,k,j] + 9*tau/(2*hy**2) * (z[i,k+1,j] - 2*z[i,k,j] + z[i,k-1,j])
            alpha_x[i] = Bx/(Cx-Ax*alpha_x[i-1])
            bet_x[i] = (F + Ax*bet_x[i-1])/(Cx - Ax*alpha_x[i-1])
        w[N-1,k] = 0
        for i in reversed(range(N-1)):
            w[i,k] = alpha_x[i]*w[i+1,k] + bet_x[i]
    for i in range(N):
        w[i,0], w[i,L-1] = 0, 0
        
    #следующий слой
    for i in range(1,N-1):
        alpha_y[0] = 0
        bet_y[0] = 0 
        for k in range(1,L-1):
            F = w[i,k] + 9*tau/(2*hx**2) * (w[i+1,k] - 2*w[i,k] + w[i-1,k])
            alpha_y[k] = By/(Cy-Ay*alpha_y[k-1])
            bet_y[k] = (F + Ay*bet_y[k-1])/(Cy - Ay*alpha_y[k-1])
        z[i,L-1,j+1] = 0
        for k in reversed(range(L-1)):
            z[i,k,j+1] = alpha_y[k]*z[i,k+1,j+1] + bet_y[k]
    for k in range(L):
        z[0,k,j+1], z[N-1,k,j+1] = 0, 0
            
%matplotlib qt
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
ax.set_xlabel('y')
ax.set_ylabel('x')
ax.set_zlabel('z')
x,y = np.meshgrid(y,x)

for j in range(M):
    for i in range(N):
        for k in range(L):
            w[i,k] = z[i,k,j]
    g = np.array(w)
    surf = ax.plot_surface(x, y, g, cmap='inferno')
    plt.pause(0.001)
    surf.remove()
    
def f(x,y,t):
    return np.sin(2*y)*np.sin(3*x)*e**(-9*13*t)

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
ax.set_xlabel('y')
ax.set_ylabel('x')
ax.set_zlabel('z')

for j in range(M):
    surf = ax.plot_surface(x, y, f(x,y,t[j]), cmap='inferno')
    plt.pause(0.0001)
    surf.remove()

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
ax.set_xlabel('y')
ax.set_ylabel('x')
ax.set_zlabel('z')

for j in range(M):
    for i in range(N):
        for k in range(L):
            w[i][k] = z[i][k][j]
    g = np.array(w)
    surf = ax.plot_surface(x, y, g - f(x,y,t[j]), cmap='inferno')
    plt.pause(0.0001)
    surf.remove()
```
