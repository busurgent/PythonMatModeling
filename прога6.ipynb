```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import math

pi = 3.1415926535
N = 30
M = 80
L = 30
hx = 1/(N-1)
tau = 1/(M-1)
hy = 1/(L-1)

y = np.zeros(L)
x = np.zeros(N)
t = np.zeros(M)

for i in range(N):
    x[i] = hx * i
    
for i in range(M):
    t[i] = tau * i
    
for i in range(L):
    y[i] = hy * i

f = [[0 for k in range(L)] for i in range(N)]
z = [[[0 for j in range(M)] for k in range(L)] for i in range(N)]

for i in range(N):
    for k in range(L):
        z[i][k][0] = 5*math.cos(pi*x[i]/2)*math.cos(3*pi*y[k]/2)
        
for i in range(N):
    for k in range(L):
        z[i][k][1] = z[i][k][0] + tau**2/2*(-5*pi**2/4*math.cos(pi*x[i]/2)*math.cos(3*pi*y[k]/2)  + 2*x[i]+2*y[k])

for j in range(1,M-1):
    for i in range(1,N-1):
        for k in range(1,L-1):
            z[i][k][j+1] = ((z[i+1][k][j]-2*z[i][k][j]+z[i-1][k][j])/hx**2 + (z[i][k+1][j]-2*z[i][k][j]+z[i][k-1][j])/hy**2)*tau**2 + 2*z[i][k][j] - z[i][k][j-1] + 2*(x[i]+y[k])*tau**2
    for i in range(N):
        z[i][0][j+1] , z[i][L-1][j+1] = z[i][1][j+1] - hy*t[j+1]**2 , t[j+1]**2*(1+x[i])
    for k in range(L):
        z[0][k][j+1] , z[N-1][k][j+1] = z[1][k][j+1] - hx*t[j+1]**2 , t[j+1]**2*(1+y[k])

%matplotlib qt
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_zlim([-3,5])
y, x = np.meshgrid(x,y)

for j in range(M):
    for i in range(N):
        for k in range(L):
            f[i][k] = z[i][k][j]
    w = np.array(f)
    surf = ax.plot_surface(x, y, w, cmap='inferno')
    plt.pause(0.0001)
    surf.remove()
    

def g(x,y,t):
    k1=np.pi/2
    k2=np.pi*3/2
    w=(k1**2+k2**2)**0.5
    return (x+y)*t**2 + 5*np.cos(w*t)*np.cos(k1*x)*np.cos(k2*y)
    

for j in range(M):
    surf = ax.plot_surface(x, y, g(x,y,t[j]), cmap='inferno')
    plt.pause(0.0001)
    surf.remove()
    
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

for j in range(M):
    for i in range(N):
        for k in range(L):
            f[i][k] = z[i][k][j]
    w = np.array(f)
    surf = ax.plot_surface(x, y, w-g(x,y,t[j]), cmap='inferno')
    plt.pause(0.0001)
    surf.remove()
```
