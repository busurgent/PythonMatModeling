```python
from numpy import zeros
from matplotlib.pyplot import plot, figure, xlabel, ylabel
l = -200
u0 = 100
t0 = 0; T = 1
N = 100
dt = (T-t0)/N
ue = zeros(N+1)
ui = zeros(N+1)
t = zeros(N+1)

ue[0] = u0
ui[0] = u0


for n in range(N):
    ui[n+1] = ui[n]/(1-dt*l)
    ue[n+1] = ue[n]*(1+dt*l)
    t[n+1] = t[n] + dt

    
figure(figsize=(10,5))
xlabel('t')
ylabel('u')
plot(t, ue, '-ro')
plot(t, ui, '-bo')

```
