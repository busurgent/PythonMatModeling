```python
import numpy as np
import matplotlib.pyplot as plt

def f(u,t):
    return np.array([-100*u*(u**2-np.pi**2)*np.cos(t)/np.sqrt(1+(-100*u*(u**2-np.pi**2)*np.cos(t))**2), 1/np.sqrt(1+(-100*u*(u**2-np.pi**2)*np.cos(t))**2)])
def df(u,t):
    x1 = (100*(np.pi**2-3*u**2)*np.cos(t))/(1+10000*u**2*(np.pi**2-u**2)**2*(np.cos(t))**2)**(3/2)
    
    x2 = -(100*(np.pi-u)*u*(np.pi+u)*np.sin(t))/(1+10000*u**2*(-np.pi**2 + u**2)**2*(np.cos(t))**2)**(3/2)
            
    y1 = -(40000*u**3*(-np.pi**2+u**2)*(np.cos(t))**2+20000*u*(-np.pi**2+u**2)**2*(np.cos(t))**2)/(2*(1+10000*u**2*(-np.pi**2+u**2)**2*(np.cos(t))**2)**(3/2))
      
    y2 = (10000*u**2*(-np.pi**2+u**2)**2*np.cos(t)*np.sin(t))/(1+10000*u**2*(-np.pi**2+u**2)**2*(np.cos(t))**2)**(3/2)
    return np.matrix([[x1,x2],[y1,y2]])
    
N = 5
a = 1
norm = []
E = np.eye(2)
while N!=5*2**15:
    tau = 7*np.pi/(N-1)
    u = np.zeros((N,2))
    u[0,0] = 1
    for i in range(N-1):
        matr = E + a*tau*df(u[i,0],u[i,1])
        k = np.dot(matr,f(u[i,0],u[i,1]))
        k = np.array(k)
        u[i+1] = u[i] + tau*k

    x = u[:,1]
    y = (1/np.pi**2+(np.pi**2-1)/(np.exp(200*np.pi**2*np.sin(x))*np.pi**2))**(-1/2)
    fig, ax = plt.subplots(1,2)
    ax[0].plot(u[:,1],u[:,0])
    ax[1].plot(x,y)
    plt.show()
    norm.append(np.sqrt(tau*sum((u[:,0]-y)**2)))
    print(norm[-1])
    print(N)
    N = N * 2
```
