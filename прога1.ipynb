```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
h = 1/(N-1)

e = 2.7182818284
y = np.zeros(N)
x = np.zeros(N)
a = np.zeros(N)
b = np.zeros(N)
c = np.zeros(N)

a[0] = 1
b[0] = h**2/2

for i in range(N):
    x[i] = h*i

for i in range(1, N):
    if i == N-1:
        a[i] = 1/(2-h-2*a[i-1])
        b[i] = (-e*h**2+2*b[i-1])/(2-h-2*a[i-1])
        
    else:
        a[i] = 1/(2-a[i-1])
        b[i] = ((e**x[i])*(h**2)+b[i-1])/(2-a[i-1])
    
y[N-1] = b[N-1]

for i in reversed(range(N-1)):
    y[i] = a[i]*y[i+1]+b[i]
    
    
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y)
    
for i in range (N):
    y[i] = -1.71828 - e**x[i] + x[i]
    
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y)
```
